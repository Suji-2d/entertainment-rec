```python
rating_data = pd.read_table("./data/movie/title.ratings.tsv")
series_data=pd.merge(series_data,rating_data,on='tconst')
```

```python
series_data=series_data[series_data['averageRating'] !='\\N']
series_data['startYear']=pd.to_numeric(series_data['averageRating'])
series_data=series_data[series_data['averageRating'] >= 6]
series_data.value_counts()
```

```python
title_data = pd.read_table("./data/movie/title.akas.tsv")
title_data.info()
```

```python
title_data=title_data.drop(['ordering','attributes','isOriginalTitle','types'],axis=1)
```

```python
title_data= title_data[title_data['language']!="\\N"]
title_data=title_data.rename(columns={"titleId": "tconst", "title": "title_regional"})
```

```python
series_data=pd.merge(series_data,title_data,on='tconst')
```

```python
series_data=series_data[(series_data['numVotes']>500)]
series_data=series_data.drop(['title_regional','region','titleType'],axis=1)
```

```python
series_data=series_data[(series_data['language']=='en') | (series_data['language']=='ja')|(series_data['language']=='fr')]
series_data.language.value_counts()
```

```python
actor_data = pd.read_table("./data/movie/title.principals.tsv")
#name_data = pd.read_table("./data/movie/name.basics.tsv")

actor_data=actor_data.drop(['ordering','category','job','characters'],axis=1)

series_data.drop_duplicates(subset=['tconst'])
series_title=series_data['tconst']
len(series_title)
#series_data=series_data.drop(['ordering','category','job','characters'],axis=1)
```

```python
actor_data=actor_data[actor_data['tconst'].isin(series_title)]
```

```python
actor_data.tconst.value_counts()
```

```python

#define function to merge columns with same names together
def same_merge(x): return ','.join(x.astype(str))

#define new DataFrame that merges columns with same names together
actor_data = actor_data.groupby('tconst').apply(lambda x: x.apply(same_merge, axis=0))
```

```python
actor_data=actor_data.rename(columns={'tconst':'del','nconst':'actors'})
```

```python
series_data1=pd.merge(actor_data,series_data,on='tconst')
series_data1.tconst.value_counts()
```

```python
series_data1 = series_data1.drop(['del','primaryTitle','startYear'],axis=1)
series_data1 = series_data1.drop_duplicates(subset=['tconst'])
```

```python

profile_data = pd.read_table("./data/movie/name.basics.tsv")
profile_data = profile_data.drop(['birthYear','deathYear','primaryProfession','knownForTitles'],axis=1)
profile_data.nconst.value_counts()
```

```python
profile_data1=profile_data[(profile_data['nconst'].isin(series_data1['actors'])|profile_data['nconst'].isin(series_data1['directors'])|profile_data['nconst'].isin(series_data1['writers']))]
profile_data1.nconst.value_counts()
```
